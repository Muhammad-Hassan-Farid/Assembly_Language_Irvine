# Key and Multi Indexing

```python
data = pd.read_csv('/content/q1_rainfall_2013.csv')
data = data.set_index('Month')
data
```

```python
rain2013 = pd.read_csv('/content/q1_rainfall_2013.csv', index_col='Month')
rain2013
```

```python
rain2014 = pd.read_csv('/content/q1_rainfall_2014.csv', index_col='Month')
rain2014
```

```python
pd.concat([rain2013,rain2014],axis=1)
```

# Multi Indexing rows and columns

```python
# for row
rain1314 = pd.concat([rain2013,rain2014], keys=['2013', '2014'], axis=0)
rain1314
```

```python
rain1314.loc['2013']
```

```python
rain1314[rain1314.index.get_level_values('Month') == 'Jan']
```

```python
# for column
rain1314 = pd.concat([rain2013,rain2014], keys=['2013', '2014'], axis='columns')
rain1314
```

# Concatenating using two different dataframes with dictionary

```python
rain_dict = {
    2013 : rain2013,
    2014 : rain2014
}
rain1314 = pd.concat(rain_dict,axis='rows')
rain1314
```
